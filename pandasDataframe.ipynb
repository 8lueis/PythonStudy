```python
import pandas as pd
```

### 딕셔너리로 데이터 프레임 생성

```python
df1 = pd.DataFrame({'이름':['정우성','이명헌','최동오','김낙수'],
                    '나이':[17, 18, 18, 18],
                    '점수':[70,86,100,88]})

print(df1)
```

### 리스트로 데이터 프레임 생성 (1) : 컬럼명 지정 생성

```python
df2 = pd.DataFrame([['송태섭',17,'북산'],
                   ['정우성',17,'산왕'],
                   ['이정환',18,'해남']],
                   columns=['성명','나이','학교'])
print(df2)
```

> 데이터 프레임 행열 바꾸기

```python
# 타입 확인
print(type(df2))
```

## .loc[] 
> 명칭 기반 인덱싱 및 슬라이싱

```python
#행 한 개 (인덱스 2번 행)
cctvDf.loc[2]
```

```python
#복수 행
cctvDf.loc[[2,5,7]]
```

```python
#슬라이싱 
cctvDf.loc[1:4]
```

```python
#행, 열 접근
cctvDf.loc[3,'소계']
#인덱스 3번 행이고 '소계' 값
```

```python
#복수 행, 열 접근
cctvDf.loc[[1,2,5],['기관명','2015년']]
```

```python
#행, 열 슬라이싱
cctvDf.loc[1:4, '2014년':]
#인덱스 1~4 & 2014년부터 끝까지
```

```python
#기관명을 인덱스 값으로 변경 
cctvDf_change = cctvDf.set_index('기관명')
print(cctvDf_change)
```

```python
#위 값을 활용한 슬라이싱
cctvDf_change.loc['관악구':'동작구',['2014년','2016년']]
```

### boolean selection row 선택

```python
# 조건에 맞는 행 전체 보기
# 2016년에 292보다 큰 데이터

cctvDf[cctvDf['2016년']>292]
```

```python
#조건이 여러 개일 때
#and 얀신자 &
#or 연산자 |

con1 = cctvDf['2015년']>=200
con2 = cctvDf['2016년']>=200
cctvDf[con1 & con2]
```

```python
cctvDf[con1 | con2]
```

```python
# 조건에 맞족하는 행, 열 
cctvDf[cctvDf['2016년']>=500][['기관명','2016년']]
```

```python
#특정 문자가 포함된 열 추출 
#df['열 이름'].str.contains('찾는 문자')

cctvDf[cctvDf['기관명'].str.contains('강')]
```

```python
#isin 함수 활용
cctvDf[cctvDf['기관명'].isin(['강남구','강동구'])]
```
