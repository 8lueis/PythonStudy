# 클라우드 컴퓨팅 ⛈
> 인터넷상의 서버를 통해 데이터 저장, 콘텐츠 사용 등 IT 관련 서비스를 사용하여
**시간 및 장소의 제한 없이 필요한 만큼의 컴퓨팅 자원을 활용**할 수 있는 방식

➡ 어떤 특정 기술을 가리키는 것은 아님

* **인터넷**을 통해 어디서든 이용할 수 있는 서비스 == 클라우드
  * 언제 어디서든 사용할 수 있음 (환경, 장소의 제약 Xx)
* 특징
  * 필요에 따라 컴퓨팅 능력을 얼마든지 이용 가능
  * 언제든지 사용하고 싶은 만큼 서비스 이용 가능
  * 서비스는 제공하는 업체에서 전부 관리
  * 소유(클라우드 제공자)와 관리(이용자)를 분리하는 방식
  * p9 총정리 되어 있음
  * 주문형 셀프 서비스: ex. CPU n개 메모리 nGB 등등... 주문하면 제공 받음
* 키워드
  * **가상화** (논리적 개념) ➡ scale up & down 가능
  * **분산 컴퓨팅 기술** ➡ 단일 컴퓨터 및 프로세서가 아닌 여러 컴퓨터 또는 서버의 네트워크를 통해 컴퓨팅 작업을 처리하는 것을 의미
    * 인터넷 또는 클라우드 기반 네트워크를 통해 처리 워크로드를 공유하는 방식으로 작동
* 다양한 정의 有
  1. NIST: IT 자원을 필요한 만큼 빌려서 사용, 실시간 확장성 지원, 사용한 만큼 비용을 지불
  2. Gartner: 인터넷 기술을 활용해 수준 높은 확장성을 가진 자원들을 서비스로 제공하는 컴퓨팅의 한 형태
  3. TTA: 가상화와 분산 처리 기술을 기반으로 인터넷을 통해 IT 자원을 임대하고 사용한 만큼 요금을 지불하는 컴퓨팅 환경

  ➡ 언제 어디서나 필요한 만큼의 컴퓨팅 자원을 필요한 시간만큼 인터넷을 통해 활용할 수 있는 컴퓨팅 방식

### 개요
* 기업이 클라우드를 이용할 경우
  * 회사의 정보를 회사 안에 구축한 시스템에서 처리하는 게 아니라, 클라우드 사업자의 데이터 센터 안의 시스템에서 처리 == IT 자산을 소유하는 게 아니라 **서비스로 이용**하는 모델
  * 클라우드 이용자는 인터넷에 접속한 후 서비스를 이용 & 또 어떻게 처리되고 있는지는 모름

* 모바일 시장의 성장
* 저렴해진 네트워크
* 기업 사용자
  * 유연한 서비스 설계와 이용 (Scale UP & Down) 과제 有
  * 구축 및 운용 부담의 경감 과제 有
  * IT 투자 비용의 저감 有
* 클라우드 사업자
  * 기업 사용자에게 컴퓨터 자원을 셀프 서비스의 형태로 제공하게 되므로 서비스 제공의 효율성이 높아짐

# Service Model ⭐
1. SaaS (Software as a Service)
2. PaaS (Platform as a Service)
3. IaaS (Infrastructure as a Service)

<hr>


## 1️⃣ SaaS (Software as a Service)
* 처음부터 끝까지 다 서비스 빌림  
* 下 Hardware - OS - Middleware - Application 上

**특징**
* 소프트웨어의 업데이트 작업은 기업 사용자가 아니라 클라우드 사업자가 수행 (즉 책임은 클라우드 사업자에게 有)
* 모든 기능을 라이센스별로 구매 XX 필요한 기능을 필요한 기간만큼 임대하는 방식

<hr>

## 2️⃣PaaS (Platform as a Service)
* Hardware - OS - Middleware 까지만 서비스 빌림

**특징**
* 기업의 어플리케이션 **실행 환경** 및 어플리케이션 **개발 환경**을 제공
* 이용자는 코딩만 하면 되..
* 자사 개발 응용 프로그램 가동 가능 == 미리 준비된 개발 환경에서 사용하고 싶은 것을 선택해 어플리케이션 개발 가능

<hr>

## 3️⃣ IaaS (Infrastructure as a Service)
* Hardware만 빌림

**특징**
* 사용자는 물리적 서버를 구입하는 게 아니라 필요할 때 필요한 만큼의 가상 서버를 만들 수 있는 것
* 대표 예시 AWS Elastic Compute Cloud (EC2) 등 有 (유연한 클라우드 컴퓨팅)
