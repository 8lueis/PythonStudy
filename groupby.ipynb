# DataFrame
## group by
* 아래 세 단계를 적용해 데이터를 그룹화 (SQL의 group by와 개념적으로 동일, 사용법은 유사)
    * 데이터 분할
    * operation 적용
    * 데이터 병함

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\juhaj\\OneDrive\\문서\\2023 김하경\\python\\edu\\3차\\pandas\\data\\train.csv')
```

### group by의 groups 속성
* 각 그룹과 그룹에 속한 index를 dict 형태로 표현

```python
class_group = df.groupby('Pclass')
```

```python
class_group
```

```python
gender_group = df.groupby('Sex')
gender_group.groups
```

### groupping 함수
* 그룹 데이터에 적용 가능한 통계 함수 (NaN은 제외 후 연산)
* count: 데이터 개수
* sum: 데이터 합
* mean, std, var: 평균, 표준편차, 분산
* min, max: 최소, 최대

```python
class_group.count()
```

```python
class_group.mean(['Survived'])
```

* index가 있는 경우 groupby 함수에 level 사용 가능
    * level은 index의 depth를 의미, 가장 왼쪽부터 0부터 증가
* set_index 함수
    * column 데이터를 index 레벨로 변경
* reset_index 함수
    * 인덱스 초기화

```python
df.head()
```

```python
df.set_index(['Pclass','Sex']).reset_index()
```
